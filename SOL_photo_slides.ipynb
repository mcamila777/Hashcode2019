```python
import numpy as np
import random


#--------------- SELECT DESIRED DATASET -------------------

#file_name = 'a_example.txt'
#file_name = 'b_lovely_landscapes.txt'
file_name = 'c_memorable_moments.txt'
#file_name = 'd_pet_pictures.txt'
#file_name = 'e_shiny_selfies.txt'

#---------------------------------------------------------
```

```python
#read data
data_full = np.loadtxt(file_name,dtype = str ,delimiter="\n", skiprows=0)

#header
header = data_full[0]
#data
data = data_full[1:]
```

```python
#convert the txt into a np.array
data_array =  np.empty([int(header), 3] ,dtype= object)

for row in range(int(header)):
    list_row = data[row].split(' ')
    data_array[row, 0] = list_row[0]
    data_array[row, 1] = list_row[1]
    data_array[row, 2] = list_row[2:]

#print (data_array)
    
```

```python
#merge pair of vertical photos into one. 
#Create pairs by joining the first half with the second half from the initial array

list_A = data_array[:,0]
index_hor = np.where( list_A == 'H')
index_vert = np.where( list_A == 'V')

#merge vertical
amount_vert = np.shape(index_vert)[1]
len_merg_vert = int(amount_vert/2)

merge_vert = np.empty([len_merg_vert, 2],  dtype= int)
merge_vert[:,0] = index_vert[0][0:len_merg_vert]
merge_vert[:,1] = index_vert[0][len_merg_vert:]

```

```python
#Get unique merged tags for merged vertical pictures
merge_vert_tags = np.empty([np.shape(merge_vert)[0], 1] ,dtype= object)

#tags for vertical merged 
for row in range(np.shape(merge_vert)[0]):
    [temp1, temp2] = merge_vert[row,:]
    first_list = data_array[temp1,2]
    second_list = data_array[temp2,2]
    
    resultList= list(set(first_list) | set(second_list))
    merge_vert_tags[row,0] = resultList
```

```python
#concatenate vertical and horizontal IDs and tags into a single array

numer_slides = np.shape(index_hor)[1] + np.shape(merge_vert)[0]
print('numer_slides')
print(numer_slides)

complete_data = np.empty([numer_slides, 2] ,dtype= object)

row = 0

if(np.shape(index_hor)[1] != 0):
    #concatenate horizontals
    for row in range(np.shape(index_hor)[1]):        
        complete_data[row, 0] = index_hor[0][row]
        complete_data[row, 1] = data_array[index_hor[0][row],2]
        
    row = row + 1

current_row = row 
    
if ( np.shape(merge_vert)[0] != 0 ):
    #concatenate vertical
    for row in range(np.shape(merge_vert)[0]):
        complete_data[row+current_row, 0] = merge_vert[row]
        complete_data[row+current_row, 1] = merge_vert_tags[row][0]
```

```python
np.shape(complete_data)[0] 
```

```python
#search  common tags in complete_data array

organized_array = np.empty((0,2) ,dtype= object)

initial_row = 0

while(np.shape(complete_data)[0] != 0):
    
    tags = complete_data[initial_row,1]
    #select a random tag from the list of tags for the current row
    rand_tag = random.choice(tags)
    
    #x = [tag_list for tag_list in complete_data[:,1] if rand_tag in tag_list]
    
    #get all the indexes for which there is a common tag with the selected rand_tag
    indexs = [indexs for indexs in range(np.shape(complete_data)[0]) if rand_tag in complete_data[indexs,1]]
    
    #store those rows into a new sorted array according to similarity 
    organized_array = np.append(organized_array, complete_data[indexs,:] , axis = 0)
    
    #delete those rows from original array in order to avoid reapiting pictures in the final array
    complete_data = np.delete(complete_data, indexs , axis=0)


  
```

```python
#Create output file

file_out_name = 'output/' + file_name

f= open(file_out_name,"w")

f.write("%d\n" % (numer_slides))

if ( np.shape(organized_array)[0]  != 0 ):
    for i in range(np.shape(organized_array)[0]):
        
        if type(organized_array[i,0]) is np.ndarray:
            print('vert')
            f.write("%d %d\n" % (organized_array[i,0][0],organized_array[i,0][1]))
        else:
            print('hor')
            f.write("%d\n" % (organized_array[i,0]))

        
```
